### Importing Required Libraries

We start by importing the necessary libraries. `matplotlib.pyplot` is the primary module for creating plots.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta

# Ensure plots are displayed inline in the notebook
%matplotlib inline
```

### Example 1: Sales Trend Over Time

This example demonstrates how to create a line chart to visualize sales trends over time.

```python
def plot_sales_trend(dates: list[str], sales: list[float]) -> None:
    """
    Args:
        dates (list[str]): List of dates.
        sales (list[float]): Corresponding sales values.
    """
    plt.figure(figsize=(7, 4), dpi=100)
    plt.plot(dates, sales, marker='o', linestyle='-', color='b', label='Sales')
    plt.title('Sales Trend Over Time', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Sales ($)', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

# Example data
number_of_days = random.randint(5, 10)
span_of_days = random.randint(number_of_days, 30)
# Generate random dates spread across an interval between number_of_days and 30 days
dates = [(datetime.now() - timedelta(days=span_of_days - i * (span_of_days // (number_of_days - 1)))).strftime('%Y-%m-%d') for i in range(number_of_days)]
sales = [random.randint(1200, 2800) for _ in range(number_of_days)]

plot_sales_trend(dates, sales)
```

### Example 2: Product Sales Distribution

This example demonstrates how to create a bar chart to compare sales across different products.

```python
def plot_product_sales(products: list[str], sales: list[float]) -> None:
    """
    Args:
        products (list[str]): List of product names.
        sales (list[float]): Corresponding sales values.
    """
    plt.figure(figsize=(7, 4))
    plt.bar(products, sales, color='skyblue', edgecolor='black')
    plt.title('Product Sales Distribution', fontsize=16)
    plt.xlabel('Products', fontsize=12)
    plt.ylabel('Sales ($)', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

items = ['T-Shirts', 'Jeans', 'Sneakers', 'Hats', 'Socks', 'Jackets', 'Sweaters', 'Shorts', 'Dresses', 'Skirts']
random.shuffle(items)

# Generate random sales data for the products
products = items[:random.randint(4, len(items))]
sales = [random.randint(1000, 5000) for _ in products]

plot_product_sales(products, sales)
```

### Example 3: Sales Contribution by Region

This example demonstrates how to create a pie chart to visualize the contribution of different regions to total sales.

```python
def plot_sales_by_region(regions: list[str], sales: list[float]) -> None:
    """
    Args:
        regions (list[str]): List of region names.
        sales (list[float]): Corresponding sales values.
    """
    plt.figure(figsize=(4, 4))
    plt.pie(sales, labels=regions, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.title('Sales by Region', fontsize=16)
    plt.tight_layout()
    plt.show()

# Example data
regions = ['North', 'South', 'East', 'West']
sales = [random.randint(1000, 5000) for _ in regions]

plot_sales_by_region(regions, sales)
```

### Example 4: Correlation Between Marketing Spend and Sales

This example demonstrates how to create a scatter plot to analyze the relationship between marketing spend and sales.

```python
def plot_marketing_vs_sales(marketing_spend: list[float], sales: list[float]) -> None:
    """
    Args:
        marketing_spend (list[float]): List of marketing spend values.
        sales (list[float]): Corresponding sales values.
    """
    plt.figure(figsize=(8, 4))
    plt.scatter(marketing_spend, sales, color='green', edgecolor='black', alpha=0.7)
    plt.title('Marketing Spend vs Sales', fontsize=16)
    plt.xlabel('Marketing Spend ($)', fontsize=12)
    plt.ylabel('Sales ($)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    # plt.xlim(0, max(marketing_spend) * 1.1)
    # Calculate and display the correlation coefficient (R^2)
    correlation_matrix = np.corrcoef(marketing_spend, sales)
    r_squared = correlation_matrix[0, 1] ** 2
    plt.text(max(marketing_spend) * 0.95, max(sales) * 0.95, f'R = {r_squared:.2f}', fontsize=10, color='red', ha='right')

    plt.tight_layout()
    plt.show()

# Example data
iterations = random.randint(5, 10)
marketing_spend = [random.randint(1000, 5000) for _ in range(iterations)]
sales = [random.randint(1000, 5000) for _ in range(iterations)]

plot_marketing_vs_sales(marketing_spend, sales)
```
